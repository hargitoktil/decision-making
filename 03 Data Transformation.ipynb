### Load library and data set

```python
# load pandas library
import pandas as pd

# load csv file
automobile = pd.read_csv('automobile_data.csv', index_col=0)
#automobile = pd.read_csv('automobile_data.csv')
```

```python
automobile.head()
```

```python
automobile.isna().sum().sum()
```

```python
automobile.fillna(automobile.median(), inplace=True)
```

### Artithmetic operation

```python
# define a new column
automobile['hpam'] = automobile['horsepower'] * automobile['average-mileage']
# print dataframe
automobile.head()
```

```python
# define a new column
automobile['new_price'] = automobile['price'] * 0.90 
# print dataframe
automobile.head()
```

### Aggregating with groupby

```python
# single reference
automobile.groupby('company').mean()
```

```python
automobile.groupby(['company', "body-style"]).mean()
```

```python
automobile.groupby(['company', "body-style"])[["price", "new_price"]].min()
```

```python
automobile.groupby('company').mean()
```

```python
automobile.groupby(['company', 'body-style']).mean()
```

```python
automobile.groupby(['company'])['price'].max()
```

```python
automobile.groupby(['company'])['price'].count()
```

```python
automobile.groupby(['company']).agg({'price': max, 
                                   'average-mileage': 'mean'})
```

### Discretization

```python
# discretization
pd.cut(automobile['price'], 3).head()
```

```python
# create price label
labels=["cheap", "standard", "expensive"]
# discretization
automobile['price_lab'] = pd.cut(automobile['price'], 3, labels=labels)
automobile.head()
```

```python
# create price label
labels=["very cheap", "cheap", "standard", "expensive", "very expensive"]
# discretization
automobile['price_lab'] = pd.cut(automobile['price'], 5, labels=labels)
automobile.head()
```

```python
# create price label
labels=["small", "middle", "high"]
# define range
ranges = [13,20,35,47]
# discretization
automobile['am_lab'] = pd.cut(automobile['average-mileage'], ranges, labels=labels)
automobile.head()
```
