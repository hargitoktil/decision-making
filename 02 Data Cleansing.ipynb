### Load library and data set

```python
# load pandas library
import pandas as pd

# load csv file
url = 'http://bit.ly/2kJwy5N'
imdb = pd.read_csv(url)
```

```python
imdb.head()
```

```python
imdb.shape
```

### Check existing NaN values

```python
#imdb.isna().head()
```

### Count NaN values

```python
imdb.isna().sum()
```

```python
imdb.isna().sum().sum()
```

### Drop records with NaN values

```python
clean_imdb1 = imdb.dropna(how='any', axis=0)
#clean_imdb1 = imdb.dropna(how='all')
```

```python
print('records number before dropping NaN values: {}'.format(imdb.shape[0]))
print('records number after dropping NaN values: {}'.format(clean_imdb1.shape[0]))
```

```python
clean_imdb1.isna().sum().sum()
```

### Drop records with all values is NaN

```python
clean_imdb2 = imdb.dropna(how='all')
```

```python
print('records number before dropping NaN values: {}'.format(imdb.shape[0]))
print('records number after dropping NaN values: {}'.format(clean_imdb2.shape[0]))
```

### Replace NaN values by 5

```python
clean_imdb3 = imdb.fillna(5)
```

```python
clean_imdb3.head()
```

### Replace NaN values by median 

```python
clean_imdb4 = imdb.fillna(imdb.median())
```

```python
clean_imdb4.head()
```

```python
# interpolation -> time series
# 5,4,3,5,4, NaN=(4+5)/2 ,5,...
imdb_2 = imdb.copy()
imdb_2.interpolate(method ='linear', inplace=True)
imdb_2.head()
```

```python
drop_dup_imdb = imdb.drop_duplicates()
```
