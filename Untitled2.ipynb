```python
imdb[["Genre", "Actors", "Rating"]]
```

```python
# row 20-31, column Director and Rating
# .loc[row, col] 
#imdb.loc[20:22, "Description":"Actors"]
#print(imdb.columns.values)
imdb.iloc[20:23, 3:6]
```

```python
# selection by condition
# select movie with the released year > 2010
condition = imdb["Year"] > 2010
imdb_2010 = imdb[condition]
imdb_2010.head()
```

```python
# selection by condition
# select movie with the released year > 2010 
# and Director name is Ridley Scott
condition = (imdb["Year"] > 2010) & (imdb["Director"] == "Ridley Scott")
imdb_2010_rs = imdb[condition]
imdb_2010_rs.head()
```
