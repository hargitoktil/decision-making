### Creating a connection

```python
# load sqlite3 library
import sqlite3

# create a connection to 
connection = sqlite3.connect('soccer.sqlite')

# create an interface to the database 
# that allows running anySQL query on our database.
cursor = connection.cursor()
```

### Get a lists of tables

```python
for row in cursor.execute("SELECT name FROM sqlite_master WHERE type='table';"):
    print(row)
```

### Reading SQL database as pandas dataframe

```python
# load pandas library
import pandas as pd

# read player table
player_table_180 = pd.read_sql_query("SELECT * FROM Player WHERE height > 200", connection)
```

```python
player_table_180.head()
```

### Reading SQL database with condition

```python
height_180 = pd.read_sql_query("SELECT * FROM Player WHERE height >= 180 ", connection)
```

```python
height_180.head()
```

### Reading another table from the database

```python
# load pandas library
import pandas as pd

# read player attributes table
player_attributes_table = pd.read_sql_query("SELECT * FROM Player_Attributes", connection)
```

```python
player_attributes_table.head()
```

### Performing inner join operation

```python
outer_join = pd.read_sql_query('SELECT * FROM Player LEFT OUTER JOIN \
                                Player_Attributes ON \
                                Player.player_api_id = Player_Attributes.player_api_id;', connection)
```
